## Configuration of Environment

```python
ENVIRONMENT_NAME = "highway-v0"
```

```python
env = gymnasium.make(ENVIRONMENT_NAME, autoreset=True, render_mode='human')
```

## Configuration of Agent

```python
model_dqn_sb3 = stable_baselines3.DQN(
    'MlpPolicy',
    env,
    policy_kwargs=dict(net_arch=[256, 256]),
    learning_rate=5e-4,
    buffer_size=15000,
    learning_starts=200,
    batch_size=32,
    gamma=0.8,
    train_freq=1,
    gradient_steps=1,
    target_update_interval=50,
    verbose=1,
    tensorboard_log="highway_dqn/"
)
```

```python
iteration_val = int(2e4)
```

```python
model_dqn_sb3.learn(iteration_val)
```

```python
model_dqn_sb3.save(f"highway_dqn/model_{iteration_val}")
```

## Evaluation of Agent

```python
model_dqn_sb3 = stable_baselines3.DQN.load(f"highway_dqn/model_{iteration_val}")
```

```python
obs, info = env.reset()
done = truncated = False
while not (done or truncated):
    action = model_dqn_sb3.predict(obs, deterministic=True)[0]
    obs, reward, done, truncated, info = env.step(action)
    env.render()
```
